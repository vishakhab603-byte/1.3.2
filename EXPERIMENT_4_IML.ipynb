```python
import pandas as pd

df1 = pd.read_csv('/content/drive/MyDrive/IML_Experiment4/Exp-4_2 diabetes.csv')
df1.head()
```

```python
df1.info()
df1.describe()
```

```python
X = df1.drop('Outcome', axis=1)
y = df1['Outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
y_prob = log_model.predict_proba(X_test)[:, 1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
```

```python
auc_score = roc_auc_score(y_test, y_prob)
print("AUC Score:", auc_score)
```

```python
plt.plot(fpr, tpr)
plt.plot([0,1], [0,1])   # Random line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_model_scaled = LogisticRegression(max_iter=1000)
log_model_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred_scaled = log_model_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

print("New Accuracy:", accuracy_score(y_test, y_pred_scaled))
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
y_prob_scaled = log_model_scaled.predict_proba(X_test_scaled)[:, 1]
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_prob_scaled)
```

```python
import matplotlib.pyplot as plt

plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_scaled)

plt.imshow(cm)
plt.title("Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100]
}

grid = GridSearchCV(
    LogisticRegression(max_iter=1000),
    param_grid,
    cv=5
)

grid.fit(X_train_scaled, y_train)

print("Best C:", grid.best_params_)
```

```python
best_model = grid.best_estimator_

y_pred_best = best_model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score
print("Best Accuracy:", accuracy_score(y_test, y_pred_best))
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(
    best_model,
    X_train_scaled,
    y_train,
    cv=5
)

print("Cross Validation Accuracy:", cv_scores.mean())
```
